```python
import pandas as pd
```

Importando a base de dados

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
dados = pd.read_csv(url2, sep=';')
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']

```

Removendo imóveis comercias

```python
df = dados.query('@imoveis_comerciais not in Tipo')
```

Selecionando apenas imóveis do tipo apartamento

```python
df = df.query('Tipo == "Apartamento"')
```

Visualizando a base de dados

```python
df.head()
```

Desafio 1 - Calcular a média de quartos por apartamento

```python
df['Quartos'].mean()
```

Desafio 2 - Conferir quantos bairros únicos existem na nossa base de dados

```python
df['Bairro'].nunique()
```

Desafio 3 - Analisar quais bairros possuem a média de valor de aluguel mais elevadas

```python
df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False)
```

Desafio 4 - Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.

```python
df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head()
```

```python
alugueis_elevados = df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False).head()
alugueis_elevados.plot(kind='barh', figsize=(14, 10), color ='blue')
```
