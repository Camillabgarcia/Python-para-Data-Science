**Acessando os dados pelo Google Planilhas**

https://docs.google.com/spreadsheets/d/1lzq0k-41-MbbS63C3Q9i1wPvLkSJt9zhr4Jolt1vEog/edit?usp=sharing

Toda planilha possui um ID para identificá-la. Este ID está presente nessa URL, entre /d/ e /edit?usp=sharing. No caso, o ID da planilha é o seguinte:

```python
sheet_id = '1lzq0k-41-MbbS63C3Q9i1wPvLkSJt9zhr4Jolt1vEog'
```

Para acessar a planilha no Colab, precisaremos alterar alguns trechos da URL. Em uma nova célula, criaremos uma variável chamada url_4, cujo valor inicial será a URL de compartilhamento até o /d/.

Para acessar a variável sheet_id nessa URL, empregaremos uma f-string, uma maneira interessante de formatar strings em Python. Basta envolver a URL entre aspas simples e incluir a letra "f" no início.

A seguir, adicionaremos alguns parâmetros para permitir o acesso à planilha. Vamos colocar uma barra após {sheet_id} e inserir o parâmetro gviz/tq. Para ter acesso à planilha, é necessário acessar a API (responsável por essa comunicação)de visualização de dados do Google.

Após gviz/tq, adicionaremos um ponto de interrogação e o parâmetro tqx=out:csv

&sheet = todos os dados da planilha serão retornados, incluindo cabeçalho, linhas e colunas.

```python
url_4 = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

Como definimos que o retorno será em formato CSV, chamaremos a função read_csv()

```python
dados_co2_sheets = pd.read_csv(url_4)
```

```python
dados_co2_sheets.head()
```

Agora, temos um hospital que deseja acessar os dados de pacientes, mas os arquivos estão todos no formato JSON. Dessa forma, é difícil extrair informações relevantes a partir dos dados. O nosso objetivo será usar a biblioteca Pandas para ler e manipular esses arquivos.

Primeiro fiz o upload do arquivo que está no meu computador e após isso copiei seu caminho para trazer pra cá.

Fazendo a leitura de arquivos no formato JSON

```python
dados_pacientes = pd.read_json('/content/pacientes.json')
```

```python
dados_pacientes
```

Deu tudo certo com o arquivo de pacientes, mas nem sempre é tão simples fazer a leitura de um arquivo JSON. Algumas vezes, as informações podem estar aninhadas dentro do arquivo JSON.

Para entendermos isso melhor, carregaremos outro arquivo usando o mesmo caminho, escolheremos o arquivo chamado "pacientes_2".

```python
dados_pacientes_2 = pd.read_json('/content/pacientes_2.json')
```

```python
dados_pacientes_2
```

O resultado é uma tabela com três colunas. A primeira coluna, intitulada "Pesquisa", tem as mesmas informações em todas as linhas, ou seja, o texto "Principais Indicadores de Doença Cardíaca". A segunda coluna exibe o ano em que a pesquisa foi feita e a última coluna, chamada "Pacientes", agrupou todas as informações de cada paciente em uma única célula.

Com isso, não conseguimos obter a separação de cada dado, tampouco ler todas as informações disponíveis, uma vez que o conteúdo da célula está cortado (conseguimos ver o ID e a faixa etária).

Este é um caso típico de informações aninhadas. Não há separação das colunas e das linhas de maneira a facilitar a leitura e a manipulação dos dados.

Normalizando os dados, transformando tudo o que estiver aninhado em uma única célula em diferentes linhas e colunas, utilizando a função json_normalize():

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
```

```python
df_normalizado
```

Salvando o arquivo normalizado

```python
df_normalizado.to_json('historico_pacientes_normalizado.json')
```
