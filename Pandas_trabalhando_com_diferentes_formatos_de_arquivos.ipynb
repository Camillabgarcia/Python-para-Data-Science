Para acessar a variável sheet_id nessa URL, empregaremos uma f-string, uma maneira interessante de formatar strings em Python. Basta envolver a URL entre aspas simples e incluir a letra "f" no início.

Agora, temos um hospital que deseja acessar os dados de pacientes, mas os arquivos estão todos no formato JSON. Dessa forma, é difícil extrair informações relevantes a partir dos dados. O nosso objetivo será usar a biblioteca Pandas para ler e manipular esses arquivos.

Primeiro fiz o upload do arquivo que está no meu computador e após isso copiei seu caminho para trazer pra cá.

Fazendo a leitura de arquivos no formato JSON

```python
dados_pacientes = pd.read_json('/content/pacientes.json')
```

```python
dados_pacientes
```

Deu tudo certo com o arquivo de pacientes, mas nem sempre é tão simples fazer a leitura de um arquivo JSON. Algumas vezes, as informações podem estar aninhadas dentro do arquivo JSON.

Para entendermos isso melhor, carregaremos outro arquivo usando o mesmo caminho, escolheremos o arquivo chamado "pacientes_2".

```python
dados_pacientes_2 = pd.read_json('/content/pacientes_2.json')
```

```python
dados_pacientes_2
```

O resultado é uma tabela com três colunas. A primeira coluna, intitulada "Pesquisa", tem as mesmas informações em todas as linhas, ou seja, o texto "Principais Indicadores de Doença Cardíaca". A segunda coluna exibe o ano em que a pesquisa foi feita e a última coluna, chamada "Pacientes", agrupou todas as informações de cada paciente em uma única célula.

Com isso, não conseguimos obter a separação de cada dado, tampouco ler todas as informações disponíveis, uma vez que o conteúdo da célula está cortado (conseguimos ver o ID e a faixa etária).

Este é um caso típico de informações aninhadas. Não há separação das colunas e das linhas de maneira a facilitar a leitura e a manipulação dos dados.

Normalizando os dados, transformando tudo o que estiver aninhado em uma única célula em diferentes linhas e colunas, utilizando a função json_normalize():

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
```

```python
df_normalizado
```

Salvando o arquivo normalizado

```python
df_normalizado.to_json('historico_pacientes_normalizado.json')
```

```python
pd.read_json('/content/historico_pacientes_normalizado.json')
```

O time de dados de uma Instituição Financeira possui um arquivo CSV que contém os dados de todos os clientes cadastrados. Porém, o que eles querem é pegar esse arquivo, colocar em um banco de dados local e a partir dele realizar consultas SQL. Assim será possível analisar dados, obter insights e tomar decisões estratégicas.

```python
import sqlalchemy
```

 A primeira importação será create_engine, que serve para criar o motor, ou seja, a conexão com o banco de dados que o fará funcionar.

Na mesma linha, inserimos uma vírgula e a importação MetaData. Essa é uma classe que representa os meta dados do banco de dados, isso inclui informações sobre as tabelas, colunas e outros objetos.

Também importamos a classe Table que representa uma tabela em um banco de dados e é usada para manipular dados em uma tabela específica.

Por fim, importamos a função inspect, que permite inspecionar a estrutura do banco de dados. Em seguida, executamos a célula.

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

Criando o Bando de Dados

```python
engine = create_engine('sqlite:///:memory:')
```

Colocando o arquivo CSV dentro do banco de dados:


*   Escrevendo em banco de dados




```python
url_5 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
import pandas as pd
```

Fazendo a leitura do arquivo utlizando a função read_csv()

```python
dados_banco = pd.read_csv(url_5)
```

```python
dados_banco
```

Para colocar as informações dos clientes da instituição financeira, foi utilizado a função .to_sql(o nome da tabela que será criada, motor)

```python
dados_banco.to_sql('clientes', engine, index=False)
```

Utilizando a função inspect() para inspecionar o banco de dados.

```python
inspector = inspect(engine)
```

Printando o nome das tabelas que temos dentro do banco de dados.

```python
print(inspector.get_table_names())
```

Acessando as informações por meio de consultas SQL.

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

Para ler uma consulta em SQL criaremos a variável empregados = pd.read_sql().

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

Para salvar como uma nova tabela os clientes que estão empregados.

```python
empregados.to_sql('empregados', con=engine, index=False)
```

Utilizamos o read_sql para lermos uma consulta, mas e se quisermos ler uma tabela inteira como a que acabamos de salvar?

Para isso, utilizamos a função pd.read_sql_table(). Dentro do parêntese inserimos aspas simples, seguido do nome da tabela que queremos executar 'empregados' e engine.

```python
pd.read_sql_table('empregados', engine)
```

Especificando as colunas a ser lidas.

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

Atualizando um banco de dados

```python
query_2 = 'SELECT * FROM clientes'
```

Lendo uma consulta.

```python
pd.read_sql(query_2, engine)
```

**Removendo informação**

Suponhamos que a pessoa de ID 5008804 cancelou sua conta na instituição financeira. Isso significa que ela não faz mais parte do quadro de clientes.

Foi feito um contexto with. Isso garante que a conexão com o banco de dados seja fechada corretamente após a execução da consulta. Depois, usamos o método engine.connect(), seguido de as conn.

```python
from sqlalchemy import text
```

```python
import sqlalchemy
```

```python
sqlalchemy.__version__
```

```python
query_3 = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(query_3)
```

Verificando se o cliente foi removido.

```python
pd.read_sql_table('clientes', engine)
```

**Atualizar uma informação**

O cliente 5008808 está com o grau de escolaridade como Ensino médio. Porém, ele fez faculdade e deseja mudar para Ensino superior.

```python
query_4 = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query_4)
```
