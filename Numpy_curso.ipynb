Ajustando uma reta para entender a taxa de crescimento do preço das maçãs.

Primeiramente, criaremos uma célula de texto onde colocaremos y=ax+b, que é a equação da reta.

```python
x = datas
```

```python
y = 2*x + 80
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
```

Note que nos é retornado o gráfico com a linha de variação dos preços e uma reta laranja que parece não corresponder com os dados do gráfico. Podemos, então, pensar em como fazer para que ela se mantenha próxima aos dados.

O primeiro passo pode ser verificar a diferença entre a reta e os dados, ou seja, analisar se a reta se ajusta aos dados. Para isso, podemos fazer uma subtração entre os dados de ambos:

```python
Moscow - y
```

O array possui diversos valores negativos. Mas se a reta estivesse abaixo da linha de preços, teríamos valores positivos.Para lidar com este problema de valores negativos e positivos, podemos elevá-los ao quadrado utilizando a função power(), que eleva valores a uma potência. Basta passarmos o valor a ser elevando e a potência desejada.

```python
np.power(Moscow - y, 2)
```

Como retorno, temos um array com os valores elevados ao quadrado. Mas note que são valores muito grandes, com várias casas decimais, dificultando a análise deste ajuste.

O ideal, portanto, é conseguir resumi-los em um único número, somando-os, por exemplo. Para isso, utilizaremos a função sum(), envolvendo o cálculo como parâmetro:

```python
np.sum(np.power(Moscow -y, 2))
```

Função sqrt - Retorna a raiz quadrada dos valores

```python
np.sqrt(np.sum(np.power(Moscow -y, 2)))
```

Existe uma fórmula para calcularmos os coeficientes angular e linear da reta, que nos permitem calcular o y e ajustar esta reta. O coeficiente angular pode ser obtido pela seguinte equação:

Calculando o coeficiente angular

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

Precisaremos calcular o quadrado de cada elemento de um array.

```python
X**2
```

Para nos certificarmos que temos a mesma quantidade do array X, basta envolvermos essa conta entre parênteses e adiconar o método .shape. Como retorno, temos 87, exatamente o número de elementos do array X.

```python
(X**2).shape
```

Voltando ao cálculo do coeficiente angular (a) :

```python
a = (n*np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n*np.sum(X**2) - np.sum(x)**2)
```

Calculando o coeficiente linear (b) :

```python
b = np.mean(Y) - a * np.mean(X)
```

Voltando a equação da reta y=ax+b :

```python
y = a * X + b
```

Vamos verificar o y utilizando a função .linalg.norm():

O processo que realizamos é uma regressão linear. y = reta

Nosso coeficiente angular é um número com vírgulas, então precisamos gerar um número do tipo float. Para isso, usamos a função random.uniform(), que gera valores aleatórios do tipo float. Ao executar, temos como retorno um array de 100 valores aleatórios do tipo float.

Esta função também recebe os parâmetros low, high e size:

```python
np.random.uniform(low=0.10, high=0.90, size=100)
```

Feito isso, teremos uma série de coeficientes angulares estimados, portanto, adicionaremos este comando a uma variável denominada coef_angulares:

```python
coef_angulares = np.random.uniform(low=0.10, high=0.90, size=100)
```

Com os coeficientes angulares armazenados na variável, calcularemos o valor da norma para cada um deles. Ou seja, testaremos cada um.

Para isso, usaremos um loop for que deve percorrer por esses coeficientes e verificar a diferença entre o array de Moscow e o array calculado com o coeficiente de teste. Para essa verificação, usaremos linalg.norm(), que deve calcular o valor de y para cada iteração:

```python
norma2 = np.array([])
for i in range(100):
  norma2 = np.append(norma2, np.linalg.norm(Moscow - (coef_angulares[i] * X +b)))
```

Ainda assim, não são números totalmente aleatórios, pois partem de um cálculo que parte de uma semente, então se utilizarmos a mesma semente conseguiremos gerar a mesma sequência de números aleatórios.

Esta semente é um número definido que será utilizado para uma sequência de cálculos da função geradora de números aleatórios da NumPy. Utilizando o mesmo número conseguiremos gerar a mesma sequência.

Para isso, precisamos utilizar uma função da NumPy, chamada random.seed(), que seleciona um número semente fixo para inicializar o gerador aleatório, optaremos por 16.

```python
np.random.seed(16)
np.random.uniform(low=0.10,high=0.90,size=100)
```
