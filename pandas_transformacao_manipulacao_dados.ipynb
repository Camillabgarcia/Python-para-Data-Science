Temos uma visualização de como é o nosso arquivo, na qual percebemos que os dados estão aninhados em formato de dicionário. Para estudá-los e identificar onde podemos prepará-los, tratá-los e transformá-los, precisamos que estejam destrinchados.

utilizaremos o método json_normalize() que fará exatamente o que precisamos: a normalização desses dados — neste caso, da coluna info_moveis.

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

Anteriormente, conhecemos nossos dados e observamos que em algumas colunas temos dados agrupados em listas. Para facilitar o acesso à esses dados a fim de que possamos trabalhá-los, precisamos desagrupá-los.

**Coletando os nomes das colunas**

Listagem dos nomes de todas as colunas do dataframe.

```python
colunas = list(dados.columns)
colunas
```

**Desagrupando os dados**

Para desagrupar esses dados, podemos utilizar o método .explode(), que separa as informações em linhas. Ele requer que enviemos o nome das colunas que serão transformadas, então o primeiro passo é coletar esses nomes.

```python
dados = dados.explode(colunas[3:])
dados
```

**Resetendo os índices para ordenar os dados.**

Para isso usaremos o método reset_index() passando os parâmetros inplace=True e drop=True. O primeiro substitui os índices no dataframe, já o segundo trata-se da criação de uma coluna com os índices antigos, porém, sendo True, não teremos essa criação.

```python
dados.reset_index(inplace=True, drop=True)
dados.head()
```

**Verificando os tipos de dados.**

É possível notar as características dos dados do nosso dataframe: há textos e números inteiros e float. Mas precisamos garantir que estejam com a tipagem correta para que possamos utilizá-los em tratamentos futuros.

Para isso, usaremos o método .info(), que nos retorna os tipos de dados.

```python
dados.info()
```

Temos várias colunas do tipo object, a transformação da tipagem precisa ser feita manualmente, pois para trabalharmos com estatística, por exemplo, precisamos que os dados numéricos sejam corretamente tipados como números.

**Tratando dados numéricos: convertendo para tipo inteiro**

Usaremos o método .astype(), que converte colunas para um tipo específico de dado. Mas, primeiro, importaremos a biblioteca NumPy, que nos fornecerá um tipo de dado (inteiro e float) para podermos converter as colunas numéricas no nosso dataframe.

Alterando o tipo:

```python
import numpy as np
```

```python
dados['max_hospedes'].astype(np.int64)
```

 Atribuição desses dados para a coluna "max_hospedes":

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

Para certificar que os dados foram corretamente substituídos, foi solicitado a tipagem das colunas:

```python
dados.info()
```

Lista com as colunas cujos dados devem ser do tipo inteiro :

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

Aplicando o mesmo processo realizado para "max_hospedes" :

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

**Tratando dados numéricos: convertendo para tipo float**

A coluna "avaliacao_geral" deve ser do tipo float, então foi feito um processo semelhante, mas agora convertendo para float64:

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

Verificando se as colunas foram alteradas :

```python
dados.info()
```

O .astype() vai funcionar em dados com cifrão ($) e vírgula (,) nas colunas "taxa_deposito", "taxa_limpeza" e "preco" são do tipo float porque correspondem a valores de moeda?

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

Focando no tratamento da coluna "preco" que, por conter caracteres como cifrão e vírgula, não pode ser convertido para o tipo float.

```python
dados['preco']
```

Cada elemento dessa coluna é uma string, então podemos fazer uma função que passe linha por linha para remover os caracteres.

Para isso, usaremos o método .apply() com lambda, que aplica uma função específica a cada elemento, no caso, a remoção dos caracteres. Feito isso, usaremos o método .replace() para substituir os caracteres de cifrão e vírgula por nada, ou seja, por uma string vazia. Depois, usaremos .strip() para remover qualquer caractere antes ou depois do elemento.

```python
dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

Atribundo as alterações ao dataframe original:

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

Usando astype depois do tratamento :

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

Conversão de duas colunas : "taxa_deposito" e "taxa_limpeza".

```python
dados[['taxa_deposito', 'taxa_limpeza']]
```

Quando tratamos da conversão de um dataframe, no qual passaremos por cada elemento, é mais viável utilizarmos o método .applymap():

```python
dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

Atribundo as alterações ao dataframe original:

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

Usando astype depois do tratamento :

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```
